```python
import pandas as pd 
df=pd.read_csv("vehicle_maintenance_data.csv")
df.drop(["Maintenance_History","Reported_Issues","Transmission_Type","Last_Service_Date","Warranty_Expiry_Date","Insurance_Premium","Mileage","Owner_Type","Brake_Condition","Battery_Status"], axis =1, inplace = True)
df.Vehicle_Model.unique()
df.Vehicle_Model.replace(['Truck', 'Van', 'Bus', 'Motorcycle', 'SUV', 'Car'],[1,2,3,4,5,6], inplace=True)
df.Fuel_Type.unique()
df.Fuel_Type.replace(['Electric', 'Petrol', 'Diesel'],[1,2,3], inplace = True)
df.Tire_Condition.unique()
df.Tire_Condition.replace(['New', 'Good', 'Worn Out'],[1,2,3], inplace = True)
X = df[['Vehicle_Model', 'Vehicle_Age','Fuel_Type', 'Engine_Size',
       'Odometer_Reading', 'Service_History', 'Accident_History',
       'Fuel_Efficiency','Tire_Condition']]
Y = df['Need_Maintenance']
from sklearn.linear_model import LinearRegression
model = LinearRegression( )
model.fit(X, Y)
res = model.predict([[6,2,2,1200,100000,1,5,15,3]])
print (str( round(res[0]*100 ,2)) + "%")
```
