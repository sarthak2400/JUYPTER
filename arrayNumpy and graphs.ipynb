```python
import numpy as np          #Numpy 
```

```python
a=np.array([1,2,3])
print("Array A",a)                                            #1-D ARRAY
b=np.array([[4,5,6],[6,7,8]])
print("Array B:\n",b)                                       # 2-D Array
```

```python
#  Indexing in array
print("first element of A:", a[0])
print ("first row of B:", b[0,:])
```

```python
#ARRAY OPERATIONS 
c=a+10
print( "c", c)
d=b*4                          #corresponding element addition
print( "d",d)                  #multiplication with const
```

```python
# Broadcasting 
e=a*np.array([1,2,3])
print("e",e)
```

```python
# Advanced indexing 
f=b[[0,1],[1,2]]                             #
print("Advanced indexing f",f)
```

```python
#Linear algebra (dot product )
g=np.dot(a,a)
print("Dot Product of a,a is",g)

#Inverse of matrix 
h= np.linalg.inv(np.array([[1,2],[3,4]]))
print("Inverse of matrix:\n",h)
```

```python
import pandas as pd
```

```python
# creating dataframes

data={'name':['sarthak','bagga','sharma'],
      'Age':[21,23,24],
      'city':['a','b','c']}
df=pd.DataFrame(data)
print(df)
```

```python
#slicing 
print(df.iloc[0])
print(df['name'])
```

```python
#basic operations (specific row nikalne ke liye)
filtered_df=df[df['Age']>23]
print(filtered_df)
```

```python
#merging the DataFrames
data2={'name':['abhay','sahil','shorya'],
       'occupation': ['Er','Er','Dr']}
df2=pd.DataFrame(data2)
merged_df=pd.merge(df , df2, on='name', how='right')
print(merged_df)
```

```python
import matplotlib.pyplot as plt
```

```python
#line plot
plt.figure(figsize=(8,4))      #8--length   4--breadth   (box ki length or breadth hai)
plt.plot([1,2,3],[4,5,6], label='line plot', color='blue', linestyle='--')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('line plot example')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#bar graph

plt.figure(figsize=(8,4))      #8--length   4--breadth   (box ki length or breadth hai)
plt.bar(['A','B','C'],[4,5,6], label='line plot', color='green')
plt.xlabel('catagories')
plt.ylabel('values')
plt.title('bar plot example')
plt.legend()
plt.grid(axis='y')
plt.show()
```

```python
import numpy as np
company = ['Google', 'Amazon', 'Microsoft', 'FB']
revenue = [96, 120,86,27]
profit=[40, 20, 34, 12]

xpos = np.arange(len(company))            #This helps in placing the bars at the correct positions on the x-axis when plotting the chart. in this len is 4 bcz 4 company hai array ki tarh store karva liya company ko
plt.xticks(xpos, company)                 #is used to set the labels of the x-axis ticks to the company names.
plt.bar(xpos+0.2, revenue, width=0.4,label='Revenue')              #revenue ko y se 0.2 jyada right me rakho
plt.bar(xpos-0.2, profit, width=0.4,label='Profit')                # profit ko y axis ke pass(left) me rakho
plt.xlabel('Company')
plt.ylabel('Revenue')
plt.legend()
plt.show()
```

```python
# importing libraries
import matplotlib.pyplot as plt
import numpy as np

# values of x and y axes
x = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50]
y = [1, 4, 3, 2, 7, 6, 9, 8, 10, 5]

plt.plot(x, y, 'b')
plt.xlabel('x')
plt.ylabel('y')

# 0 is the initial value, 51 is the final value
# (last value is not taken) and 5 is the difference
# of values between two consecutive ticks
plt.xticks(np.arange(0, 51, 5))
plt.yticks(np.arange(0, 11, 1))
plt.show()
```

```python
                       #PIE Graph 
```

```python
exp_vals =[1600, 600, 300, 410, 250]
exp_labels=['Home rent', 'Food', 'Phone/Internet Bill', 'Car', 'Other utilities']
plt.axis('equal')                                #mtlb gol rhe lamba or choda gol na ho simple equal rhe 
#plt.pie(values, lables)
plt.pie(exp_vals, labels = exp_labels, autopct='%0.2f%%', explode=[0,0,0.2,0.1,0], radius=1.5)           #autopct='%0.2f%% used to display the value 
plt.savefig('pie.jpeg', bbox_inches='tight',pad_inches=2)      #bbox_inches='tight' minimizes this extra space, ensuring that the saved image tightly encloses the figure and its elements.
plt.show()
                      #Pie graph 
```

```python
                   #SUBPLOTES      ( ALL THE GRAPHS TOGETHER ONCE)
fid, axs =plt.subplots(2, 2, figsize=(10,8))
axs[0,0].plot([1,2,3],[4,5,6])
axs[0,0].set_title('Subplot 1')
axs[0,1].bar(['A','B','C'],[4,5,6] , color='orange')
axs[0,1].set_title('Subplot 2')
axs[1,0].scatter([1,2,3],[8,9,24] , color='purple')
axs[1,0].set_title('Subplot 3')
axs[1,1].hist([1,2,3,4,5,2,4] , bins=3,color='blue')
axs[1,1].set_title('Subplot 4')
plt.tight_layout()
plt.show()
```
